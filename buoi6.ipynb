```python
def soNguyenTo (n:int) -> bool:
    if (n<2):
        return False
    for i in range (2,int(n**0.5)+1):
        if (n%i==0):
            return False
            print ("Xin chao cac ban") # Dòng này sẽ không hiện ? tại sao nhỉ?
    else:
        return True # phải gặp return True thì mới dừng, False thì vẫn tiếp tục
n = int(input())
for x in range (2,n+1):
    if soNguyenTo(x):
        print (x, end=" ")
```

```python
def compareBMI(w1,h1,w2,h2) -> int:
    bmi1= w1/(h1*h1)
    bmi2 = w2/(h2*h2)
    if bmi1 > bmi2:
        return 1
    elif bmi1 < bmi2:
        return 2
    else:
        return 0
assert compareBMI (58,1.58,66,1.70) == 1
print ("Pass all test")
```

```python
def map_grade (s):
    if s >= 9:
        return "A+"
    elif s >= 8.5: # ở nhánh này mặc định đã là nhỏ hơn 9 rồi, không cần so với 9
        return "A"
    elif s >= 8:
        return "B+"
    elif s > 7:
        return "B"
    else:
        return "E"

assert map_grade (9.5) == "A+"
print("Passed")
```

```python
def piecewise_mode3(n):
    if n%3 == 1:
        return 5*n+1
    elif n%3 == 2:
        return 6*n+2
    else:
        return 7*n+3
```
