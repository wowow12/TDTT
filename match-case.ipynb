## Học Match - Case

Nhập vào một số nguyên đại diện cho tháng trong năm. In ra số ngày tương ứng tháng đó.

```python
x = int(input())
nam = int(input())
match x:
    case 1|3|5|7|8|10|12:
        print (f"Tháng {x} có 31 ngày")
    case 4|6|9|11:
        print (f"Tháng {x} có 30 ngày")
    case 2:
        print (f"Tháng {x} có 28/29 ngày")
    case _:
        print (f"Tháng {x} không tồn tại ở thế giới con người.")
```

```python
n = int(input)
match n:
    case n if n%2==0:
        print ("Số chẵn")
    case _:
        print ("Số lẻ")
```
