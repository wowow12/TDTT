### Bài 1: Tìm số lớn nhất của 1 list

```python
lst = list(map(int,input().split()))
if lst:
    max = lst[0]
for i in lst:
    if i > max:
        max = i
print (max)
```

### Bài 2b: Tìm số nhỏ nhất của 1 list và vị trí của nó, nếu có nhiều số nhỏ nhất bằng nhau thì lấy chỉ số đầu tiên (nhỏ nhất)
10 0 0 12 34 5 6 7  8 89 
0 1

```python
lst = [10,0,0,12,34,5,6,7,8,89,0,1]
nho = lst[0]
so = 1
for i,x in enumerate(lst,1): # chú ý phải là chỉ số trước giá trị
    if x < nho:
        nho = x
        so = i
print (nho, so)
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không

```python
lst = list(map(int,input().split()))
if lst == lst [::-1]:
    print (True)
else:
    print (False)
```

```python
def doiXung (lst):
    l = len(lst)
    for i in range (l//2):
        if lst[i] != lst[l-1-i]:
            return False
    return True
assert doiXung([1,2,3,4,3,2,1]) == True
print ("Passed")
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
lst = [3,4,6,7,1,9,0,0,23,0,0,34,67,89]
lst1 =[]
for x in lst:
    if x == 0:
        lst1.insert(0,x)
    else:
        lst1.append(x)
print (lst1)
```

### Bài 6: Xóa tất cả các số giống nhau trong list, Các số khác giữ nguyên vị trí trước sau

```python
lst = [3,4,6,7,1,9,0,0,23,0,0,34,67,89]
ans = []
for x in lst:
    if x not in ans:
        ans.append(x)
print (ans)
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
def daySo (lst):
    lst = [1,3,5,6,7,8,9,6]
    for i in range (0,len(lst)+1):
        if lst[i+1] > lst[i]:
            lst[i+1] = lst[i]
print (daySo(lst)) 
```
