Nhập một số nguyên đến khi được một số nguyên dương thì thôi.  Kiểm tra xem n đó có phải số nguyên tố không.

```python
import math
n = -5
while n <= 0:
    n = int(input())
    if n==1:
        print ("Khong phai so nguyen to")
    else:
        for i in range(2,int(math.sqrt(n)) + 1):
            if n%i==0:
                print ("Khong phai so nguyen to")
                break
        else:
            print ("La so nguyen to")
```
