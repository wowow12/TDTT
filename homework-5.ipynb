# W5A1. 

Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) ->
int: 

```python
def Max_Of_Two(a: int, b: int) -> int:
    """
    Hàm nhận vào hai số nguyên và trả về số lớn hơn.
    """
    if a > b:
        return a
    else:
        return b
if __name__ == "__main__":
    num1 = int(input("Nhập số thứ nhất: "))
    num2 = int(input("Nhập số thứ hai: "))
    ket_qua = Max_Of_Two(num1, num2)
    print(f"Số lớn hơn trong hai số {num1} và {num2} là: {ket_qua}")
```

# W5A2.

Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]: 

```python
from typing import Tuple

def swap(a: int, b: int) -> Tuple[int, int]:
    a, b = b, a
    return a, b

if __name__ == "__main__":
    x, y = 5, 10
    print(f"Trước khi hoán đổi: x = {x}, y = {y}")
    x, y = swap(x, y)
    print(f"Sau khi hoán đổi:  x = {x}, y = {y}")
```

# W5A3.

Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 

```python
import math
def is_prime(n: int) -> bool:
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    limit = int(math.sqrt(n))
    for i in range(3, limit + 1, 2):
        if n % i == 0:
            return False
    return True
if __name__ == "__main__":
    num = int(input("Nhập một số nguyên: "))
    if is_prime(num):
        print(f"{num} là số nguyên tố.")
    else:
        print(f"{num} không phải là số nguyên tố.")
```

# W5A4.

VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
import math
def is_perfect_number(n: int) -> bool:
    if n <= 1:
        return False
    tong_uoc = 1
    can_bac_hai = int(math.sqrt(n))
    for i in range(2, can_bac_hai + 1):
        if n % i == 0:
            tong_uoc += i
            if i != n // i:
                tong_uoc += n // i
    return tong_uoc == n
if __name__ == "__main__":
    test_number = int(input("Nhập số cần kiểm tra: "))
    
    if is_perfect_number(test_number):
        print(f"{test_number} là số hoàn hảo.")
    else:
        print(f"{test_number} không phải là số hoàn hảo.")
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 

```python
from typing import List
def find_first(nums: List[int], k: int) -> int:
    for i, val in enumerate(nums):
        if val == k:
            return i
    return -1
if __name__ == "__main__":
    my_list = [10, 20, 30, 40, 50, 30]
    target = 30
    vi_tri = find_first(my_list, target)
    if vi_tri != -1:
        print(f"Số {target} xuất hiện đầu tiên tại vị trí: {vi_tri}")
    else:
        print(f"Không tìm thấy số {target} trong danh sách.")
```

liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
